```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
Activations1 = np.load("tmp/ActivationsMnist_Layer1.npy")
Activations2 = np.load("tmp/ActivationsMnist_Layer2.npy")

print(Activations1.shape)
print(Activations2.shape)
```

```python
Norm1 = np.linalg.norm(Activations1, axis = (3, 4))
Norm1 = np.linalg.norm(Norm1, axis = (1, 2))
```

```python
print("Activations by Node:")
print(Norm1)
```

```python
print(np.argsort(Norm1))
```

```python
best_nodes = np.where(24 <= np.argsort(Norm1))[0]
```

```python
print("Best Nodes by Activation:", best_nodes)
```

```python
plt.figure(1, figsize=(30, 30))
j = 1

for i in best_nodes:
    plt.subplot(2, 4, j)
    plt.title('Filter ' + str(i))
    plt.imshow(Activations1[i, 1, :, :, 0], interpolation="nearest", cmap="gray")
    j = j + 1
    
plt.show()
```

```python
plt.figure(1, figsize=(30, 30))
j = 1

for i in best_nodes:
    plt.subplot(2, 4, j)
    plt.title('Filter ' + str(i))
    plt.imshow(Activations1[i, 3, :, :, 0], interpolation="nearest", cmap="gray")
    j = j + 1
    
plt.show()
```

# Image action by random node:

```python
subset = random.sample(range(32), 2)
print("Selected nodes:", subset)

Pictures_Activations1 = Activations1[subset,]
print("Shape:", Pictures_Activations1.shape)
```

```python
Pictures_Norm1 = np.linalg.norm(Pictures_Activations1, axis = (3, 4))
Pictures_Norm1 = np.linalg.norm(Pictures_Norm1, axis = 2)
print(Pictures_Norm1.shape)
```

```python
print("Sample activations:", Pictures_Norm1[0, :10])
print("Ranks:", np.argsort(Pictures_Norm1[0, :10]))
```

```python
best_pictures = np.where(92 <= np.argsort(Pictures_Norm1[0,]))[0]
print("Best activations:", best_pictures)
```

```python
plt.figure(1, figsize=(30, 30))
j = 1

for i in best_pictures:
    plt.subplot(2, 4, j)
    plt.title('Image ' + str(i))
    plt.imshow(Activations1[subset[0], i, :, :, 0], interpolation="nearest", cmap="gray")
    j = j + 1
    
plt.show()
```

```python
best_pictures2 = np.where(92 <= np.argsort(Pictures_Norm1[1,]))[0]
print("Best activations:", best_pictures2)
```
