```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#
# read original images
#
eval_data_dmf = np.load("tmp/DMF/eval_data_DMF.npy")
eval_data_dmf.shape
```

```python
#
# get filter numbers for each of the layers
#
filtersLayer1 = np.load("tmp/DMF/RandomFiltersIndex_Layer1.npy")
filtersLayer2 = np.load("tmp/DMF/RandomFiltersIndex_Layer2.npy")
filtersLayer3 = np.load("tmp/DMF/RandomFiltersIndex_Layer3.npy")
print(filtersLayer1)
print(filtersLayer2)
print(filtersLayer3)
```

```python
#
# read activations
#

# layer 1
Activations1 = np.load("tmp/DMF/ActivationsDMF_Layer1_Features.npy")

# layer 2
Activations2 = np.load("tmp/DMF/ActivationsDMF_Layer2_Features.npy")

# layer 3
Activations3 = np.load("tmp/DMF/ActivationsDMF_Layer3_Features.npy")

print(Activations1.shape)
print(Activations2.shape)
print(Activations3.shape)
```

```python
#
# read deconvolved activations
#

# layer 1
DeconvActivations1 = np.load("tmp/DMF/ActivationsDMF_Layer1.npy")

# layer 2
DeconvActivations2 = np.load("tmp/DMF/ActivationsDMF_Layer2.npy")

# layer 3
DeconvActivations3 = np.load("tmp/DMF/ActivationsDMF_Layer3.npy")

print(DeconvActivations1.shape)
print(DeconvActivations2.shape)
print(DeconvActivations3.shape)
```

```python
#
# get index of images that maximally activates each the specific nodes of each layer
#
maxIndexesLayer1 = np.load("tmp/DMF/BestImagesDMF_Layer1.npy")
maxIndexesLayer2 = np.load("tmp/DMF/BestImagesDMF_Layer2.npy")
maxIndexesLayer3 = np.load("tmp/DMF/BestImagesDMF_Layer3.npy")
print(maxIndexesLayer1)
print(maxIndexesLayer2)
print(maxIndexesLayer3)
```

# Best Nodes Layer 1

##  Node 15

```python
plt.figure(figsize = (30,30))
# index
iNode = 0
indexNode1 = filtersLayer1[0]

# images
plt.subplot(331)
plt.imshow( eval_data_dmf[int( maxIndexesLayer1[iNode, 0] ),:,:,:] , interpolation="nearest" )
plt.subplot(332)
plt.imshow( eval_data_dmf[int( maxIndexesLayer1[iNode, 1] ),:,:,:] , interpolation="nearest" )
plt.subplot(333)
plt.imshow( eval_data_dmf[int( maxIndexesLayer1[iNode, 2] ),:,:,:] , interpolation="nearest" )

# feature map visualization
plt.subplot(334)
plt.imshow( Activations1[iNode,0,:,:,0] )
plt.subplot(335)
plt.imshow( Activations1[iNode,1,:,:,0] )
plt.subplot(336)
plt.imshow( Activations1[iNode,2,:,:,0] )

# feature map visualization
plt.subplot(337)
plt.imshow( DeconvActivations1[iNode,0,:,:,:] )
plt.subplot(338)
plt.imshow( DeconvActivations1[iNode,1,:,:,:] )
plt.subplot(339)
plt.imshow( DeconvActivations1[iNode,2,:,:,:] )

plt.savefig('../img/node{}_layer1_DMF'.format(indexNode1))
plt.show()
```
